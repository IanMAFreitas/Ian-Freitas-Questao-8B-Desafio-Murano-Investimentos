```python
def Q8_B(stocks):
  import matplotlib.pyplot as plt
  import pandas as pd
  import seaborn as sns
  import numpy as np

  stocks_df = pd.DataFrame()

  for stock in stocks:
    stock_ = stock + '.us.txt'
    stocks_df[stock.upper()] = pd.read_csv(stock_)['Close']

  corr = stocks_df.corr()

  mask = np.zeros_like(corr)
  triangle_indices = np.triu_indices_from(mask)
  mask[triangle_indices] = True

  plt.figure(figsize=(10,8))  
  sns.heatmap(corr, cmap='vlag', mask = mask, annot=True, vmin=-1, vmax=1)

stocks_list = ['aapl','elec','fss','ghy','googl','mgu','mjco','msft','ndsn','shak']

Q8_B(stocks_list)
```
